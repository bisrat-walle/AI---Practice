```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
df
```

```python
sns.pairplot(df)
```

```python
df.describe().round(2)
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = scaler.fit_transform(df)
```

```python
df = pd.DataFrame(df, columns=["Id", "SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"])
df
```

```python
sns.pairplot(df)
```

```python
df.describe().round(2)
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
df = scalar.fit_transform(df)
df = pd.DataFrame(df, columns=["Id", "SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"])
df
```

```python
df.describe().round(2)
```

```python
sns.pairplot(df)
```
