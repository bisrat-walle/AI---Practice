```python
# drawing sin graph
x = np.arange(0, 4, 0.01)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
plt.subplot(2, 1, 1)
y = np.tan(x)
plt.plot(y)

plt.subplot(2, 1, 2)
y = np.cos(x)
plt.plot(y)
plt.show()
```

```python
plt.style.use("seaborn-v0_8-dark")
plt.figure(figsize=(10, 6), dpi=100)

x = np.linspace(-np.pi, np.pi, 100, endpoint=True)
s = np.sin(x)
c = np.cos(x)

plt.plot(x, s, color="blue", linewidth=1, label="blue", linestyle="-")
plt.plot(x, c, color="red", linewidth=.8, label="purple", linestyle="-.")

plt.legend()
plt.xlim(-4, 4)
plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
plt.ylim(-1, 1)
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))
plt.savefig("exercise.png", dpi=55)
plt.show()
```

```python
x = np.linspace(0, 10, 5)
y = np.linspace(0, 10, 5)

plt.scatter(x, y, c="green", s=80, marker=".")
plt.show()
```
